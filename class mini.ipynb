```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Brain_GSE50161.csv')
```

```python
df_melted = df.melt(id_vars=['samples', 'type'], var_name='Gene', value_name='Expression_Level')

```

```python
df_heatmap = df_melted.pivot_table(index='Gene', columns='type', values='Expression_Level', aggfunc='mean')
```

```python
plt.figure(figsize=(10, 15)) # Adjust size as needed
sns.heatmap(df_heatmap, cmap='viridis', cbar_kws={'label': 'Average Expression Level'})
plt.title('Average Gene Expression Heatmap by Sample Type')
plt.xlabel('Sample Type')
plt.ylabel('Gene')
plt.tight_layout()
plt.show()
```
